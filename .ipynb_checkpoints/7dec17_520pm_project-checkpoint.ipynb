### General Properties

> **Loading in the 'tmdb-movies.csv' file, and printing out a few lines, and checking the total length. **

**Checking for NaNs and Zeros in rating column**

```python
#to check for NaN condition
def isnan(f):
    return isinstance(f, float) and f!=f

def iszero(q):
    if isnan(q):
        return False
    if q ==0:
        return True
    else:
        return False
    

#check if vote_average has NaN or 'zero' values

assert len(df[df.vote_average.apply(isnan)]) == 0
assert len(df[df.vote_average.apply(iszero)]) == 0
```

No NaNs or Zeros in the vote_average(rating) column

### Data Cleaning (Performing cleaning steps)

>- Deleting duplicate row 
>- Splitting single row data with multiple genres into multiple rows with single genres. 

>Add yearbin column
- a: 1960 - 1969
- b: 1970 - 1979
- c: 1980 - 1989
- d: 1990 - 1999
- e: 2000 - 2009
- f: 2010 - 2015

**Deleting duplicate row. Keeping first occurence. Doing operation on the original dataframe. **

```python
#delete duplicate row
df.drop_duplicates(inplace=True)
len(df)  #new length - 10865
```

The duplicate row has been deleted. 

**Splitting single row data with multiple genres into multiple rows with single genres.**

```python
s = df['genres'].str.split('|').apply(pd.Series, 1).stack() #splitting by the delimiter '|', and stacking all of them up. 
s.index = s.index.droplevel(-1) #Dropping the extra level of index
del df['genres'] #dropping original 'genres' column 
s.name = 'genres' # for join to work, adding name
newdf = df.join(s)
newdf.genres.head()
```

We see that we've successfully split multiple genre rows into single genre rows, preserving original indexing

**Adding new column for year ranges. 10 year intervals. [ 1960-1969, 1970-1979, ... and so on ]**

```python
#adding new column for year ranges


newdf['year_bin'] = pd.cut(newdf.release_year, [1960,1969,1979,1989,1999,2009,2019], \
                           include_lowest=True, labels=['a','b','c','d','e','f'] )
```

```python
newdf.year_bin.describe() #Checking if everything checks out
```

The new column 'year_bin' has been successfully added. with lablels a,b,c,e,d,f for 10 year intervals from 1960 to 2020

**Change all zeros to NaN values in budget_adj and revenue_adj columns**

```python
#Change all zeros to NaN values in budget_adj and revenue_adj
newdf.budget_adj = newdf.budget_adj.replace(0, np.nan)
newdf.revenue_adj = newdf.revenue_adj.replace(0, np.nan)

#Check if zeros exist

print len(newdf[newdf.revenue_adj.apply(iszero)])
print len(newdf[newdf.budget_adj.apply(iszero)])
```

Rows with zero values can be movies where budget/revenue data was not available. 
Zeros are made to NaNs above so that, when doing simple arithmetic calculation with these values, if value is zero (now NaN), the resulting value will also be NaN. 
 

**Adding profit_adj column --> Profit = ( Revenue - Budget ) **

```python
#Adding new column 'profit_adj'
newdf['profit_adj']=newdf.revenue_adj.sub(newdf.budget_adj, axis=0)

# replace 0 values with nans in profit_adj and roi_adj
newdf.profit_adj = newdf.profit_adj.replace(0, np.nan)
```

New column 'profit_adj' was added, and values were calculated and stored. 
Rows with zero values in the 'profit_adj' column was changed to NaN; on the assumption that this was some kind of data error, where profit and revenue matched up exactly .

## Exploratory Data Analysis


>### Which are the most popular, and least popular genres based on the number of movies in each genre in our database

```python
print topgenres
```

The top genres are Drama, Comedy, Thriller, Action, Romance, Horror and Aventure.

```python
print leastgenres
```

The least popular genres are Western, TV Movie, Foreign, War and History

>### How have user ratings and profits for movies in popoular genres, changed over 10 year periods. Keeping in mind the the average rating of movies falling in the studied genres in the 10 year periods.  

**Creating new dataframe containing only the top 7 genres which we are analysing**

```python
studied_df = newdf[newdf.genres.isin(topgenres.index)]
studied_df.genres.describe()
```

Stored new dataframe as studied_df, and checked values in 'genre' column. Everything checks out. 

**Grouping by genres, and finding average rating for them from 1960-2015**

```python
groupby_genres = studied_df.groupby('genres')
avg_rating_studied_genres = groupby_genres['vote_average'].mean()
print avg_rating_studied_genres
```

We see the average rating of the most popular genres across the entire study period. Interesting! 

**Grouping by year_bin, and finding average rating of genres studied in each yearbin**

```python

avg_rating_yearbins = studied_df.groupby('year_bin')['vote_average'].mean()
print avg_rating_yearbins
```

We see the average rating of studied genres in each yearbin. Interesting! We see a gradual decline in the ratings. 
Maybe because the number of mediocre movies being produced in recent times have gone high, or maybe because the public is more critical, or maybe because the samples in our database is not right. So many possibilities. Interesting nonetheless!


**Using pivot_table on the dataframe, so that we get average rating of each genre in each 10 year interval being analysed. **

```python
ratings_table = studied_df.pivot_table(index='genres',columns='year_bin')['vote_average']
print ratings_table
```

The values in the table is not normalized to any baseline

**Plotting a line graph of the ratings as it is.**

```python
plot1 = ratings_table.transpose().plot(kind='line')
plot1.set(xlabel='Year Bins', ylabel='Average Ratings')
plot1.set(title='Average Ratings per genre in 10 year intervals')
```

**LEGEND for x- axis**
>a : 1960 - 1969  b : 1970 - 1979  c : 1980 - 1989  d : 1990 - 1999  e : 2000 - 2009  f : 2010 - 2019

We see the non-standardised average rating values across the 10 year interval year bins in graphic. Clearly 'horror' genre movies are not at all doing good. 

**Standardising the ratings_table with average rating of genres studied in each yearbin. (avg_rating_yearbins). 
Plotting the same **


```python
std_table1 = ratings_table - avg_rating_yearbins
print std_table1

plot2 = std_table1.transpose().plot()
plot2.set(xlabel='Year Bins', ylabel='Standardised Average Ratings')
plot2.set(title='Average Ratings per genre in 10 year intervals - standardised plot 1')
```

**LEGEND for x- axis**
>a : 1960 - 1969  b : 1970 - 1979  c : 1980 - 1989  d : 1990 - 1999  e : 2000 - 2009  f : 2010 - 2019

It looks like Comedy, Adventure, and Romance, have been rated higher, compared to other top movie genres, in recent years. 

>### How have the user ratings for popular genres changed over these periods, keeping in mind the average user sentiment for each genre over 50 years.

**Standardising the ratings_table with average rating of movies per genre in the study period. (avg_rating_studied_genres). 
Plotting the same **


```python
std_table2 = ratings_table.sub(avg_rating_studied_genres, axis=0)
print std_table2

plot3 = std_table2.transpose().plot()
plot3.set(xlabel='Year Bins', ylabel='Standardised Average Ratings')
plot3.set(title='Average Ratings per genre in 10 year intervals - standardised plot 2')
```

**LEGEND for x- axis**

We see almost same trends even here. We observe that Comedy, Romance and Adventure movies, have begun to get better ratings than what they used to get before, in the recent years. 

>###  How have the profits for movies in popular genres, changed over 10 year periods in our database. 

**Using pivot_table on the dataframe,  getting mean of profits_adj column of movies in each popular genres, in each yearbin. **

```python
profits_table = studied_df.pivot_table(index='genres', columns = 'year_bin')['profit_adj']
print profits_table
```

The profits_table looks right. 

**Plotting the table**

```python
profits_plot = profits_table.transpose().plot()
profits_plot.set(xlabel='Year Bins', ylabel='Average Profit (unit : 10^8 USD)')
profits_plot.set(title='Average Profits per genre in 10 year intervals')
```

**LEGEND for x- axis**
>a : 1960 - 1969  b : 1970 - 1979  c : 1980 - 1989  d : 1990 - 1999  e : 2000 - 2009  f : 2010 - 2019

We observe that adjusted profits were way higher from around 1960 to 1990 for almost all genres. 
This maybe because less movies were produced in that time period. Demand was high, and supply was less. So people flocked to what they have. 
Also, this maybe because there were great advancements in these periods in movie making, and the people were amazed and wanted to experience this. 
We observe that Adventure and Action genre, has been making good money in the recent years and they are on a climb. 
The spike in Horror movie profits for the yearbin b: 1970-1979 is intriguing. 

**Checking horror movie with high profit**

```python
for i, k in studied_df.groupby('genres'):
    if i == 'Horror':
        print k[k.profit_adj.isin(k.profit_adj.nlargest(5))]
```

We see that out of the top five horror movies with highest profits, three of them are from yearbin b (1970 - 1979). And two of them made profits in billions. The three being : Jaws 1, Jaws2, I am Legend
